This is a simple Data visualization project using python and the pandas library. A csv file of the states in the USA and their respective sizes in sq.mi is imported with pandas into a dataframe. 

```python
# YOUR CODE HERE AND INSERT ADDITIONAL CELS AS NEEDED.

# https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-areas.csv

import pandas as pd

state = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-areas.csv', delimiter=',')

print state
```

The total and the average area, the state with the minimum and maximum area is determined using pandas Dataframe methods

```python
Total = state['area (sq. mi)'].sum()

Average = state['area (sq. mi)'].mean()

min = state['area (sq. mi)'].min()
max = state['area (sq. mi)'].max()

print "The total area of all states in the US is",Total,"sq. mi"
print "The average of all areas of states in the US is",Average,"sq. mi"

print "The state with the least area is", state['state'][state['area (sq. mi)'].idxmin()],"with an area of",min,"sq. mi"
print "The state with the largest area is", state['state'][state['area (sq. mi)'].idxmax()],"with an area of",max,"sq. mi"
```

We make A pie plot to create a proportional representation of the area of states in America using pandas  

```python
plot = state.plot.pie(y='area (sq. mi)', labels = state['state'],figsize=(30, 30))
```

We make A bar plot that represents categorical data with rectangular bars with lengths proportional to the values that they represent. 
One axis of the plot shows states, and the other axis represents the area

```python
ax = state.plot.bar(x='state', y='area (sq. mi)', rot=0,figsize=(30, 30))
```
